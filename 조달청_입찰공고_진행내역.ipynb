```python
import pandas as pd

# 조달청 로우 데이터 로드 및 전처리
file_path = 'UI-ADOFAA-010R.입찰공고 내역.csv'

try:
    # 상단 메타데이터 제외하고 로드 (encoding: utf-16, delimiter: tab)
    # 5492라인 부근 파싱 에러 방지를 위해 on_bad_lines='skip' 처리
    df = pd.read_csv(file_path,
                     encoding='utf-16',
                     sep='\t',
                     skiprows=10,
                     on_bad_lines='skip')

    # Header 위치 안 맞을 경우 재조정
    if '공고명' not in df.columns and '입찰공고명' not in df.columns:
        df = pd.read_csv(file_path,
                         encoding='utf-16',
                         sep='\t',
                         skiprows=11,
                         on_bad_lines='skip')

    # 데이터 로드 확인
    print(f"Total Rows: {len(df)}")
    print(df.columns.tolist())

    # 시각화 툴(Tableau 등) 연동을 위해 utf-8-sig로 재저장
    df.to_csv('cleaned_bidding_data.csv', index=False, encoding='utf-8-sig')
    print("Export complete: cleaned_bidding_data.csv")

except Exception as e:
    print(f"Error: {e}")
```

```python
import pandas as pd

file_path = 'UI-ADOFAA-010R.입찰공고 내역.csv'

# 데이터 시작 지점(Header Row) 검색
with open(file_path, 'r', encoding='utf-16') as f:
    lines = f.readlines()

target_idx = 0
for i, line in enumerate(lines):
    # '공고' 및 '기관' 키워드 매칭으로 헤더 인덱스 탐색
    if '공고' in line and '기관' in line:
        target_idx = i
        break

# 지정된 인덱스 기반 데이터 로드 및 전처리
# Parse error 방지를 위해 on_bad_lines skip 처리
df = pd.read_csv(file_path,
                 encoding='utf-16',
                 sep='\t',
                 skiprows=target_idx,
                 on_bad_lines='skip')

# Tableau 연동용 데이터 추출 (UTF-8-SIG 인코딩)
output_name = 'cleaned_final.csv'
df.to_csv(output_name, index=False, encoding='utf-8-sig')

print(f"Process Complete. Header Index: {target_idx}")
print(f"Columns: {df.columns.tolist()}")
```
